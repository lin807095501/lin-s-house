```python
import matplotlib.pyplot as plt
import numpy as np

train_loss = []
f = open('result-cifar10.txt', 'r')
for line in f:
    datas = line.split('\t')
    if datas[0].find('[390/391]') != -1:
#         print(datas)
        loss = datas[3].split(' ')
        avg_loss = loss[2][1:-1]
#         print(avg_loss)
        train_loss.append(float(avg_loss))
        

x = [i for i in range(len(train_loss))]
print(x)
print(train_loss)
# # print(x)
plt.plot(x, train_loss, color='r')
# print(train_loss[0])
# print(train_loss[-1])
plt.xlabel("epoch")
plt.ylabel("train loss")
plt.title("The train loss on cifar10 dataset")
#显示图例
plt.savefig("cifar10_train_loss.jpg", dpi=300)
plt.show()
```
