```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import rcParams

config = {
            "font.family": 'serif',
            "font.size": 15,# 相当于小四大小
            "mathtext.fontset": 'stix',#matplotlib渲染数学字体时使用的字体，和Times New Roman差别不大
            "font.serif": ['Times New Roman'],#宋体
            'axes.unicode_minus': False # 处理负号，即-号
         }
rcParams.update(config)

#准备绘制数据
x = [1, 2, 3, 4, 5, 6]
y = [58.03, 57.67, 57.29, 56.78, 57.23, 56.70]
# z = [57.68, 58.34, 57.36, 56.92, 57.01, 57.12]
# w = [71.58, 72.18, 71.59, 71.12, 71.15, 70.94]
# z = [52.92, 49.43, 49.16, 48.97, 49.04, 48.94]
# w = [45.06, 36.20, 30.92,21.40, 12.71, 10.51]
# "g" 表示红色，marksize用来设置'D'菱形的大小
plt.plot(x, y, marker='s', markersize=7, label="Diginetica")
# plt.plot(x, z, marker='^', markersize=7, label="Retailrocket")
# plt.plot(x, w, marker='o', markersize=7, label="Yoochoose 1/64")
# plt.plot(x, w, color="g", marker='o', markersize=7, label="SR-GNN")
#绘制坐标轴标签
# plt.title("Diginetica")
# plt.xlabel("The number of layers", fontsize=12)
plt.ylabel("HR@20(%)")
# 坐标轴刻度字体大小
# plt.tick_params(labelsize=14)

# 设置y坐标取值范围
plt.ylim([54, 62])

#显示图例
plt.legend(loc=0)
#设置网格线
plt.grid()
#调用 text()在图像上绘制注释文本
#x1、y1表示文本所处坐标位置，ha参数控制水平对齐方式, va控制垂直对齐方式，str(y1)表示要绘制的文本
for x1, y1 in zip(x, y):
    plt.text(x1, y1, str(y1), ha='center', va='bottom')
# for x1, z1 in zip(x, z):
#     plt.text(x1, z1, str(z1), ha='center', va='bottom')
# for x1, z1 in zip(x, w):
#     plt.text(x1, z1, str(z1), ha='center', va='bottom')
#保存图片
plt.savefig("DIG_gMLP_HR@20.pdf", dpi=300)
plt.show()
```
