```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import rcParams

config = {
            "font.family": 'serif',
            "font.size": 13,# 相当于小四大小
            "mathtext.fontset": 'stix',#matplotlib渲染数学字体时使用的字体，和Times New Roman差别不大
            "font.serif": ['Times New Roman'],#宋体
            'axes.unicode_minus': False # 处理负号，即-号
         }
rcParams.update(config)

size = 3
# 返回size个0-1的随机数
max_pool = np.array([39.56, 47.27, 61.75])
concat =  np.array([39.64, 46.81, 60.95])
gate = np.array([39.85, 47.01, 60.29])
sum_pool = np.array([44.82, 50.86, 61.82])
# x轴坐标, size=5, 返回[0, 1, 2, 3]
x = np.arange(size)

# 有a/b/c/d三种类型的数据，n设置为4
total_width, n = 0.8, 4
# 每种类型的柱状图宽度
width = total_width / n 

# 重新设置x轴的坐标
x = x - (total_width - width) / 2 

# 设置xy坐标名称
plt.ylabel("HR@10(%)")

# 设置y坐标取值范围
plt.ylim([30, 70])

#F38181
#FCE38A
#EAFFD0
#95E1D3

# 画柱状图
plt.bar(x - width, concat, width=width, label="Concat", color="#F38181", edgecolor='black', linewidth=0.5, hatch='/')
plt.bar(x, gate, width=width, label="Gating", color="#FCE38A", edgecolor='black', linewidth=0.5, hatch='-')
plt.bar(x + width, max_pool, width=width, label="Max", color="#EAFFD0", edgecolor='black', linewidth=0.5, hatch='\\')
plt.bar(x + 2 * width, sum_pool, width=width, label="Sum", color="#95E1D3", edgecolor='black', linewidth=0.5, hatch='X')
# 显示图例
plt.legend(loc=2)

# 功能1
x_labels = ["Diginetica", "Retailrocket", "Yoochoose 1/64"]
# 用第1组...替换横坐标x的值
plt.xticks(x, x_labels)
plt.tick_params(labelsize=14)

# 功能2
for i, j in zip(x - width, concat):
    plt.text(i, j + 0.01, "%.1f" % j, ha="center", va="bottom")
for i, j in zip(x, gate):
    plt.text(i, j + 0.01, "%.1f" % j, ha="center", va="bottom")
for i, j in zip(x + width, max_pool):
    plt.text(i, j + 0.01, "%.1f" % j, ha="center", va="bottom")
for i, j in zip(x + 2 * width, sum_pool):
    plt.text(i, j + 0.01, "%.1f" % j, ha="center", va="bottom")

# 显示柱状图
plt.savefig(fname="./aggr_HR@10.pdf", dpi=300)
plt.show()
```
